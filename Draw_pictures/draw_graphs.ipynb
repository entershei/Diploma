```python
from graphviz import Digraph, Graph
```

```python
fg = Digraph(name='Functional graph', comment='The example of functional graph')
fg.edge_attr.update(arrowsize='0.7')
fg.attr('node', shape='circle', fixedsize='true', width='0.3')
fg.engine = 'neato'

fg.node('1', pos='0,0!')
fg.node('12', pos='0.6,0!')
fg.node('8', pos='1.2,0!')
fg.node('2', pos='1.2,-0.6!')
fg.node('3', pos='1.8,-0.6!')
fg.node('10', pos='2.4,-0.6!')
fg.node('5', pos='3,-0.6!')
fg.node('13', pos='3.6,-0.6!')
fg.node('4', pos='0,-0.6!')
fg.node('9', pos='1.2,-1.2!')
fg.node('6', pos='0.6,-1.2!')
fg.node('11', pos='2.4,0!')
fg.node('7', pos='2.4,-1.2!')


fg.edges(['41', '69', ['7', '10'], '92', ['11', '10']])
fg.edge('2', '12', color='crimson')
fg.edge('3', '3', color='crimson', headport='ne', tailport='nw')
fg.edge('5', '10', color='crimson')
fg.edge('8', '2', color='crimson')
fg.edge('10', '5', color='crimson')
fg.edge('12', '8', color='crimson')


# print(fg.source)
fg.render('functional_graph', view=False)  
fg
```

```python
t = Graph(name='Bipartite tree', comment='Tree contains only the path')
t.attr('node', shape='circle', fixedsize='true', width='0.4')
t.engine = 'fdp'
t.graph_attr.update({'splines': 'compound'})
# fdp neato

a = Graph(name='cluster_A', node_attr={'style': 'filled', 'color': 'white'})
a.graph_attr.update({'label': 'A-вершины', 'style': 'filled', 'color':'lightgrey'})

a.node('1', pos='0,3!')
a.node('2', pos='0,2.4!')
a.node('3', pos='0,1.8!')
a.node('4', pos='0,1.2!')
a.node('5', pos='0,0.6!')
a.node('6', pos='0,0!')

a.edge('2', '4', color='red2', headport='nw', tailport='sw', splines='line')
a.edge('5', '6', color='red2')

t.subgraph(a)

b = Graph(name='cluster_B', node_attr={'style': 'filled'})
b.graph_attr.update({'label': 'B-вершины'})

b.node('7', pos='2,3.6!')
b.node('8', pos='2,3!')
b.node('9', pos='2,2.4!')
b.node('10', pos='2,1.8!')
b.node('11', pos='2,1.2!')
b.node('12', pos='2,0.6!')
b.node('13', pos='2,0!')
b.node('14', pos='2,-0.6!')

b.edge('9', '10', color='mediumblue')
b.edge('10', '12', color='mediumblue', headport='ne', tailport='se')
b.edge('12', '13', color='mediumblue')
b.edge('12', '14', color='mediumblue', headport='ne', tailport='se')

t.subgraph(b)

t.edge('1', '7')
t.edge('1', '8')
t.edge('2', '8')
t.edge('3', '9')
t.edge('4', '11')
t.edge('5', '13')
t.edge('6', '11')

# print(fg.source)
t.render('bipartite_tree', view=False)  
t
```

```python
t = Digraph(name='Circles')
t.edge_attr.update(arrowsize='0.6')
t.attr('node', shape='circle', fixedsize='true', width='0.4')
t.engine = 'dot'
t.graph_attr.update({'splines': 'compound'})
# fdp neato

c0 = Digraph(name='cluster_0')
c0.node('1')

c1 = Digraph(name='cluster_1')
c1.node('2')
c1.node('8', style='filled', color='lightgrey')

c2 = Digraph(name='cluster_2')
c2.node('4')

c3 = Digraph(name='cluster_3')
c3.node('5')
c3.node('6')
c3.node('11', style='filled', color='lightgrey')

c4 = Digraph(name='cluster_4')
c4.node('12', style='filled', color='lightgrey')
c4.node('13', style='filled', color='lightgrey')

c5 = Digraph(name='cluster_5')
c5.node('14')

t.subgraph(c0)
t.subgraph(c1)
t.subgraph(c2)
t.subgraph(c3)
t.subgraph(c4)
t.subgraph(c5)

t.edge('1', '8', constraint='false')
t.edge('2', '4', constraint='false', color='red3')
t.edge('4', '11', constraint='false')
t.edge('5', '13', constraint='false')
t.edge('12', '14', constraint='false', color='mediumblue')

# print(fg.source)
t.render('pi', view=False)  
t
```
